```python
import pandas as pd 
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

#Reading the CSV and printing the head

```python

data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/Instagram.csv", encoding = 'latin1')
data = data.dropna()
# data.head()
print(data.head())
```

#Plotting scatter Linear Relationship graph using Plotly 

```python
figure = px.scatter(data_frame = data, 
                    x="Impressions",
                    y="Likes", 
                    size="Likes", 
                    trendline="ols", 
                    title = "Relationship Between Likes and Impressions")
figure.show()
```

#Plotting scatter Linear Relationship graph using Matplotlib 

```python
plt.figure(figsize=(10,8))
plt.title('Relationship Between Likes & Impressions')
sns.regplot(x="Impressions", y="Likes", data=data)
plt.show()
```
